```python
import cv2
import os
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
char="stt"
dem=0
for i in range(30):
    video_path = f"D:\\save_videos\\{i+1}.mp4"
    crop_interval = 1
    output_dir = "D:\\sudan"

    # Open the video capture object
    cap = cv2.VideoCapture(video_path)

    # Get video frame rate (might not be accurate for all videos)
    frame_rate = cap.get(cv2.CAP_PROP_FPS)

    # Calculate number of frames to skip between captures based on desired interval
    frames_to_skip = int(frame_rate * crop_interval)

    # Count frames and image number
    frame_count = 0
    while True:
        # Read a frame from the video
        ret, frame = cap.read()

        # Break the loop if there are no more frames
        if not ret:
            break

        # Check if it's time to capture a frame (multiple of frames_to_skip)
        if frame_count % frames_to_skip == 0:
            # Increase image number for unique filenames
            dem += 1

            # Save the frame as an image (adjust filename format as needed)
            image_filename = f"{output_dir}/frame_{char+str(dem)}.png"
            cv2.imwrite(image_filename, frame)
            print(f"Saved {image_filename}")

        # Increase frame count
        frame_count += 1

# Release the video capture object
cap.release()
```
